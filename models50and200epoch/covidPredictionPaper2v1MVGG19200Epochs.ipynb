```python
from tensorflow.keras.layers import Input, Flatten, Dense, Dropout, GlobalAveragePooling2D, BatchNormalization
from tensorflow.keras.models import Model
from tensorflow.keras.applications import VGG19
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical

from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import LabelBinarizer
from sklearn.metrics import classification_report
import numpy as np
import matplotlib.pyplot as plt
from glob import glob
import cv2
from datetime import datetime

# Get the current time
now = datetime.now()
print("Start Time =", now)

# Define size to which images are to be resized
IMAGE_SIZE = [224, 224]

# Training configurations
epochs = 200     # adjust as needed
batch_size = 32
n_splits = 10    # 2–10 folds

# Paths
covid_path = '/home/CT_COVID'
noncovid_path = '/home/CT_NonCOVID'

# Load files
covid_files = glob(covid_path + '/*')
noncovid_files = glob(noncovid_path + '/*')

print("COVID images:", len(covid_files))
print("Non-COVID images:", len(noncovid_files))

# --- Preprocessing function ---
def preprocess_image(image):
    gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    contrast_enhanced = clahe.apply(gray)
    contrast_enhanced_rgb = cv2.cvtColor(contrast_enhanced, cv2.COLOR_GRAY2RGB)
    denoised_image = cv2.GaussianBlur(contrast_enhanced_rgb, (3, 3), 0)
    return denoised_image

# Load COVID images
covid_images, covid_labels = [], []
for f in covid_files:
    img = cv2.imread(f)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, (224, 224))
    img = preprocess_image(img)
    covid_images.append(img)
    covid_labels.append("CT_COVID")

# Load Non-COVID images
noncovid_images, noncovid_labels = [], []
for f in noncovid_files:
    img = cv2.imread(f)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, (224, 224))
    img = preprocess_image(img)
    noncovid_images.append(img)
    noncovid_labels.append("CT_NonCOVID")

# Normalize
covid_images = np.array(covid_images) / 255.0
noncovid_images = np.array(noncovid_images) / 255.0

# Merge datasets
X = np.concatenate((covid_images, noncovid_images), axis=0)
y = np.concatenate((covid_labels, noncovid_labels), axis=0)

# Encode labels
le = LabelEncoder()
y_encoded = le.fit_transform(y)       # convert text labels to 0/1
y_bin = to_categorical(y_encoded, 2)  # one-hot encode

print("Final dataset:", X.shape, y_bin.shape)

# --- Function: Build Modified VGG19 ---
def build_modified_vgg19():
    base_model = VGG19(weights="imagenet", include_top=False, input_tensor=Input(shape=(224, 224, 3)))

    # Custom head
    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = Dense(512, activation="relu")(x)
    x = BatchNormalization()(x)
    x = Dropout(0.5)(x)
    x = Dense(256, activation="relu")(x)
    x = BatchNormalization()(x)
    x = Dropout(0.5)(x)
    outputs = Dense(2, activation="softmax")(x)

    model = Model(inputs=base_model.input, outputs=outputs)

    # Freeze base model layers
    for layer in base_model.layers:
        layer.trainable = False

    model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
    return model

# --- Cross-validation ---
skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)
fold_no = 1
all_fold_acc = []
all_histories = []

for train_idx, val_idx in skf.split(X, y):
    print(f"\n### Training Fold {fold_no}/{n_splits} ###")

    X_train, X_val = X[train_idx], X[val_idx]
    y_train, y_val = y_bin[train_idx], y_bin[val_idx]

    # Data augmentation
    train_aug = ImageDataGenerator(
        rotation_range=20,
        width_shift_range=0.2,
        height_shift_range=0.2,
        horizontal_flip=True
    )
    train_gen = train_aug.flow(X_train, y_train, batch_size=batch_size, shuffle=True)

    # Build modified model
    model = build_modified_vgg19()

    steps_per_epoch = len(X_train) // batch_size

    history = model.fit(
        train_gen,
        validation_data=(X_val, y_val),
        steps_per_epoch=steps_per_epoch,
        epochs=epochs,
        verbose=1
    )

    all_histories.append(history.history)

    # Evaluate
    scores = model.evaluate(X_val, y_val, verbose=0)
    print(f"Fold {fold_no} Accuracy: {scores[1]*100:.2f}%")
    all_fold_acc.append(scores[1])

    # Classification report
    y_pred = model.predict(X_val, batch_size=batch_size)
    y_pred_bin = np.argmax(y_pred, axis=1)
    y_val_bin = np.argmax(y_val, axis=1)
    print(classification_report(y_val_bin, y_pred_bin))

    # Plot per-fold accuracy/loss
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train')
    plt.plot(history.history['val_accuracy'], label='Validation')
    plt.title(f'Fold {fold_no} - Accuracy')
    plt.xlabel('Epoch'); plt.ylabel('Accuracy'); plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train')
    plt.plot(history.history['val_loss'], label='Validation')
    plt.title(f'Fold {fold_no} - Loss')
    plt.xlabel('Epoch'); plt.ylabel('Loss'); plt.legend()
    plt.show()

    fold_no += 1

# --- Summary across folds ---
print("\n=== Cross-validation Results ===")
print("Accuracies per fold:", [round(a*100, 2) for a in all_fold_acc])
print("Mean Accuracy: %.2f%% (+/- %.2f%%)" % (np.mean(all_fold_acc)*100, np.std(all_fold_acc)*100))

# Averaged curves
epochs_range = range(1, epochs+1)
mean_acc = np.mean([h['accuracy'] for h in all_histories], axis=0)
mean_val_acc = np.mean([h['val_accuracy'] for h in all_histories], axis=0)
mean_loss = np.mean([h['loss'] for h in all_histories], axis=0)
mean_val_loss = np.mean([h['val_loss'] for h in all_histories], axis=0)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, mean_acc, label='Train (avg)')
plt.plot(epochs_range, mean_val_acc, label='Validation (avg)')
plt.title('Average Accuracy Across Folds')
plt.xlabel('Epoch'); plt.ylabel('Accuracy'); plt.legend()

plt.subplot(1, 2, 2)
plt.plot(epochs_range, mean_loss, label='Train (avg)')
plt.plot(epochs_range, mean_val_loss, label='Validation (avg)')
plt.title('Average Loss Across Folds')
plt.xlabel('Epoch'); plt.ylabel('Loss'); plt.legend()
plt.show()

print("# Save the model to use it for prediction in future")
model.save('mvgg19.h5')

# End time
print("End Time =", datetime.now())
```
